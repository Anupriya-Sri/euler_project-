# Problem-1_Multiples of 3 or 5

If we list all the natural numbers below 10 that are multiples of 3 or 5, we get 3, 5, 6 and 9. The sum of these multiples is 23.

Find the sum of all the multiples of 3 or 5 below 1000.

```python
# array of numbers below 1000
X = np.arange(0, 1000)

# multiples of 3
X_3 = X[X%3 == 0]

# multiples of 5
X_5 = X[X%5 == 0]

# multiples of 15
X_15 = X[X%15 == 0]

# multiples of 15 are repeated in multiples of 3 and multiples of 5
# So, sum = multiples of 3 + multiples of 5 - multiple of 15
sum_multiples_3_or_5 = X_3.sum() + X_5.sum() - X_15.sum()
sum_multiples_3_or_5
```
